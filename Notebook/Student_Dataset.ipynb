```python
#importing libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

#loading dataset
df = pd.read_csv(r'c:\Users\User\OneDrive\Desktop\synthetic_student_dataset_missing.csv')
```

```python
#dispay first 5 row
print(df.head())
```

```python
#display shape
print(df.shape)
```

```python
#display information
print(df.info())
```

```python
#display stastical summary
print(df.describe())
```

```python
#check missing value
print(df.isnull().sum())
print(df.isnull().sum()/len(df)*100)
```

```python
#data cleaning of numerical columns
num_cols = df.select_dtypes(include=['number']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
print(df[num_cols].isnull().sum())
```

```python
#data cleaning of categorical values
cat_cols = df.select_dtypes(include=['string']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
print(df[col].isnull().sum())
```

```python
#feature engineering to create average score
df['Average_Score'] = (df['math_score'] + df['english_score'] + df['science_score']) / 3
print(df['Average_Score'].head())
```

```python
#feature engineering to create total score
df['Total_Score'] = df['math_score'] + df['english_score'] + df['science_score']
print(df['Total_Score'].head())
```

```python
#feature engineering of pass / fail classification
df['Result'] = df["Average_Score"].apply(lambda x: 'pass' if x >= 40 else 'fail')
print(df[['Average_Score', 'Result']].head())
print(df.value_counts('Result'))
```

```python
#label- encoding of pass / fail result
df['Result_encoded'] = df["Result"].replace({"fail": 0, "pass": 1})
print(df[['Result', 'Result_encoded']].head())
```

```python
#saved clean dataset
df.to_csv(r'c:\Users\User\OneDrive\Desktop\synthetic_student_cleaned_dataset.csv', index = False)
```

```python
#outlier detection using IQR method
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f'Outliers in {col}:')
    print(outliers[[col]])
```

```python
#distribution of average score
plt.figure(figsize=(6,3))
sns.histplot(df["Average_Score"], bins=20, kde=True)
plt.title("Distribution of Average Score")
plt.show()
```

```python
#Average score by gender
plt.figure(figsize=(6,3))
sns.barplot(x = 'gender', y = 'Average_Score', data=df)
plt.title("Average Score by Gender")
plt.show()
```

```python
#pass / Fail distribution
plt.figure(figsize=(6,3))
sns.countplot(x = "Result", data=df)
plt.title("Pass vs Fail Distribution")
plt.show()
```
